```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
img_path = 'C:/Users/HP/Downloads/HD-wallpaper-cute-pup-2-puppy.jpg'  # Provide the path to your image file
img = Image.open(img_path)

# Extract RGB channels
r, g, b = img.split()

# Display the image and its R, G, B channels
fig, axs = plt.subplots(1, 4, figsize=(20, 5))  # Setting up the subplots
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')  # Turn off axis

axs[1].imshow(r, cmap='Reds')
axs[1].set_title('Red Channel')
axs[1].axis('off')  # Turn off axis

axs[2].imshow(g, cmap='Greens')
axs[2].set_title('Green Channel')
axs[2].axis('off')  # Turn off axis

axs[3].imshow(b, cmap='Blues')
axs[3].set_title('Blue Channel')
axs[3].axis('off')  # Turn off axis

plt.show()
```
